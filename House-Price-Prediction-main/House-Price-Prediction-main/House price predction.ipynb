```python
from sklearn.model_selection import train_test_split

X=data.drop(['median_house_value'],axis=1)
y=data['median_house_value']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4)
```

```python
train_data=X_train.join(y_train)
```

```python
# Assuming train_data is your DataFrame
numeric_columns = train_data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = train_data[numeric_columns].corr()

plt.figure(figsize=(15, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu") #corr()--> to find the pairwise correlation of all columns
```

```python
train_data['total_rooms']=np.log(train_data['total_rooms']+1)#This line takes the 'total_rooms' column in your DataFrame, adds 1 to each value, and then applies the natural logarithm (base e) to the result. Adding 1 before taking the logarithm is a common practice to handle cases where the original values are zero because the logarithm of zero is undefined.
train_data['total_bedrooms']=np.log(train_data['total_bedrooms']+1)#Similar to the first line, this line applies the natural logarithm transformation to the 'total_bedrooms' column.
train_data['population']=np.log(train_data['population']+1)#This line applies the natural logarithm transformation to the 'population' column
train_data['households']=np.log(train_data['households']+1)#Similarly, this line applies the natural logarithm transformation to the 'households' column.
```

```python
train_data.hist(figsize=(15,8))
```

```python
#train_data.ocean_proximity.value_counts()#is used to count the occurrences of unique values in the 'ocean_proximity' column of the train_data DataFrame

train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1) #joined the dummies with the trained data set
```
